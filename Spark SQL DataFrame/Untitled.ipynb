```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import count, desc , col, max, struct
import matplotlib.pyplot as plt
```

```python
spark = SparkSession.builder.appName("PysparkTutorial").getOrCreate()
spark
```

```python
spotify_df = spark.read.csv("spotify-2023.csv", header=True, inferSchema=True)
spotify_df.show(10)
```

```python
spotify_df.printSchema()
```

```python
spotify_df = spotify_df.drop('artist_count')
```

```python
spotify_df.select('in_apple_playlists', 'in_apple_charts').show()
```

```python
bts_apple = spotify_df.select('track_name', 'artist(s)_name', 'in_apple_playlists').where(spotify_df['artist(s)_name'] == 'BTS')
bts_apple.show()
```

```python
spotify_df.select('artist(s)_name').filter(spotify_df['artist(s)_name'] == 'BTS').groupby('artist(s)_name').agg(count('*').alias('number_of_songs')).show(10) 
```

```python
bts_apple.groupby('artist(s)_name').agg(sum('in_apple_playlists').alias('max')).show(10) 
```
